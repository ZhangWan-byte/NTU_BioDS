# Q2.3 code simulation

```python
from tqdm import tqdm
import math
import matplotlib.pyplot as plt
```

```python
E, S, ES, P = 1, 10, 0, 0

k1, k2, k3 = 100, 600, 150

dt = 1e-6

li_E = []
li_S = []
li_ES = []
li_P = []
```

```python
for i in tqdm(range(10000000)):
    
    E1, S1, ES1, P1 = E, S, ES, P
    
    E1 = E1 - k1*E1*S1*dt + k2*ES1*dt + k3*ES1*dt
    S1 = S1 - k1*E1*S1*dt + k2*ES1*dt
    ES1 = ES1 + k1*E1*S1*dt - k3*ES1*dt - k2*ES1*dt
    P1 = P1 + k3*ES1*dt

    if E1<=1e-10 or S1<=1e-10:
        break
    
    E, S, ES, P = E1, S1, ES1, P1

    li_E.append(E)
    li_S.append(S)
    li_ES.append(ES)
    li_P.append(P)
```

```python
plt.plot(li_E)
plt.plot(li_S)
plt.plot(li_ES)
plt.plot(li_P)

plt.legend(["E","S","ES","P"])
```

```python
plt.figure(figsize=(16,12))

plt.subplot(221)
plt.title("E")
plt.plot(li_E)

plt.subplot(222)
plt.title("S")
plt.plot(li_S)

plt.subplot(223)
plt.title("ES")
plt.plot(li_ES)

plt.subplot(224)
plt.title("P")
plt.plot(li_P)
```
